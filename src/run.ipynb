# Create Dataframe

```python
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
# Let's look at original dataset
all_rick = pd.read_csv('/content/RickAndMortyScripts.csv')
all_rick.head(10)
```

```python
contexted = []

n = 7

for i in range(n, len(all_rick['line'])):
  row = []
  prev = i - 1 - n # we additionally substract 1, so row will contain current responce and 7 previous responces
  for j in range(i, prev, -1):
    row.append(all_rick['line'][j])
  contexted.append(row)
```

```python
columns = ['response', 'context']
columns = columns + ['context/'+str(i) for i in range(n-1)]
columns
```

```python
df = pd.DataFrame.from_records(contexted, columns=columns)
df.head(5)
```

```python
trn_df, val_df = train_test_split(df, test_size = 0.1)
trn_df.head()
```

```python
trn_df.to_csv('/content/trn_df.csv', index=False)
val_df.to_csv('/content/val_df.csv', index=False)
```
